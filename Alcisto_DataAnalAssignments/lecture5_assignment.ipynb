```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Load the datasettttt
utrecht_houses = pd.read_csv("utrechthousinglarge.csv")

X = utrecht_houses[['zipcode', 'lot-len', 'lot-width', 'lot-area', 'house-area', 'garden-size', 'balcony', 'x-coor']]
y = utrecht_houses['y-coor']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)

test_predictions = model.predict(X_test)
test_res = y_test - test_predictions

sns.scatterplot(x=y_test,y=test_res)
plt.axhline(y=0, color='r', linestyle='--')
```

```python
len(test_res)
```

```python
sns.displot(test_res,bins=25,kde=True)
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

fig, ax = plt.subplots(figsize=(6, 8), dpi=100)
_ = stats.probplot(test_res, plot=ax)
plt.show()
```

```python
final_model = LinearRegression()
final_model.fit(X,y)
```

```python
import matplotlib.pyplot as plt

# Assuming you have defined final_model and X
y_hat = final_model.predict(X)

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 6))

axes[0].plot(X['lot-len'], y, 'o')
axes[0].plot(X['lot-len'], y_hat, 'o', color='green')
axes[0].set_ylabel("y-coor")
axes[0].set_title("lot length")

axes[1].plot(X['lot-area'], y, 'o')
axes[1].plot(X['lot-area'], y_hat, 'o', color='brown')
axes[1].set_title("lot Area")
axes[1].set_ylabel("y-coor")

axes[2].plot(X['house-area'], y, 'o')
axes[2].plot(X['house-area'], y_hat, 'o', color='red')
axes[2].set_title("House Area")
axes[2].set_ylabel("y-coor")

plt.tight_layout()
plt.show()
```

```python
residuals = y_hat - y
sns.scatterplot(x=y,y=residuals)
plt.axhline(y=0, color='r', linestyle='--')
```

```python
final_model.coef_
coeff_df = pd.DataFrame(final_model.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

y-coordinate shows the location of the house, it is an integer value between 5000 and 6000. therefore depending on the location of the house, it shows the length, lot area, and house area of Utrecht, Netherlands.
